```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
N = 100
norm_rv = sts.norm(0, 1)
X = norm_rv.rvs(N)
```

```python
x_arr = np.arange(-10, 10, 0.05)
K = 400
mark_1 = np.array([abs(np.sum(X[0 : n + 1]) / (n + 1)) for n in range(N)])
parameters = [[0, 1], [0, 100], [10, 1], [10, 100]]

mark_21 = np.array([abs((np.sum(X[: n + 1]) + float(parameters[0][0])/ 
                         parameters[0][1]) / (n + 1 + 1. / parameters[0][1])) 
                    for n in range(N)]) 
mark_22 = np.array([abs((np.sum(X[: n + 1]) + float(parameters[1][0])/ 
                         parameters[1][1]) / (n + 1 + 1. / parameters[1][1])) 
                    for n in range(N)]) 
mark_23 = np.array([abs((np.sum(X[: n + 1]) + float(parameters[2][0])/ 
                         parameters[2][1]) / (n + 1 + 1. / parameters[2][1])) 
                    for n in range(N)]) 
mark_24 = np.array([abs((np.sum(X[: n + 1]) + float(parameters[3][0])/ 
                         parameters[3][1]) / (n + 1 + 1. / parameters[3][1])) 
                    for n in range(N)]) 
```

```python
n_arr = np.arange(0, 100, 1)
plt.figure(figsize = (15, 8))
plt.plot(n_arr, mark_1, alpha = 1)
plt.plot(n_arr, mark_21, alpha = 1)
plt.plot(n_arr, mark_22, alpha = 1)
plt.plot(n_arr, mark_23)
plt.plot(n_arr, mark_24, alpha = 0.7)
plt.ylim([0, 0.1])
plt.legend(['VM (0, theta)', 'BM (0, 1)', 'BM (0, 100)', 'BM (10, 1)', 'BM (10, 100)'], loc = 0, framealpha=0.5)
```

## Параметры (0,1), (0, 100), (10, 100) являются хорошими для априорного распределения. Для них байесовская оценка почти равняется оценке макс. правдоподобия. (10, 1) - не годятся для априорного распределения.
