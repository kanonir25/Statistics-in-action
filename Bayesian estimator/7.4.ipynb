```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
t_0 = 500
t = 100000
N = 1000
```

```python
processes = []
file = open('6.csv', 'r')
for i in range(4):
    file.readline()

for line in file:
    for value in line.split():
      processes.append(float(value))
file.close()
```

```python
stohastic_values = np.zeros(N)
stohastic_values[0] = processes[0]
for i in range(N - 2):
    stohastic_values[i + 1] = processes[i + 1] - processes[i] 
```

```python
def apri(alpha, beta): 
    n_arr = np.arange(N)
    mark = np.array([(beta + np.sum(stohastic_values[: n + 1]) / (alpha + n) ) for n in range(N)])
    plt.plot(n_arr, mark)
    plt.ylim([90, 100])
    print(mark)
```

```python

apri(1, 0)
```

```python
apri(1, 1)
```

```python
apri(0, 1)
```
