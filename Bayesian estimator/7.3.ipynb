```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

### Создадим выборку из распределения Коши и возьмем в качестве априорного для него нормальное распределение, с параметром $a = 0$, а $\sigma^{2}$ возьмем из правила трех сигм (в 0,5 содержится 0,95% информации => $\sigma^{2} = ({\frac{0.5}{3}})^2$ 

```python
N = 100
cauchy_rv = sts.cauchy(0, 1)
X = cauchy_rv.rvs(N)
mark_1 = np.array([np.sum(X[: n + 1]) / (n + 1) for n in range(N)])
a = 0
sigma_sq = (0.5 / 3) ** 2
mark_2 = np.array([(np.sum(X[: n + 1 ]) + a / sigma_sq) / (n + 1 + 1 / sigma_sq) for n in range(N)])
plt.plot(np.arange(N), abs(mark_1))
plt.plot(np.arange(N), abs(mark_2))
plt.legend(['VM', 'Bayes (' + str(a) + ',' + str(sigma_sq) + ')'])
```

### Баейсовская оценка является хорошей, если считать распределение Коши подходящей для модели, описанной в условии. Параметры априорного распрделения выбраны грамотно!)
