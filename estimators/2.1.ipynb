```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
beg = 0
teta = 1
N = 10000
unif_rv = sts.uniform(beg, teta - beg) 
X = unif_rv.rvs(N)
arr1 = np.array([np.sum(X[0 : n + 1]) / (n + 1) for n in range (N)]) * 2 
arr2 = np.array([np.sum(X[0 : n + 1]) / (n + 1) + np.max(X[0 : n + 1]) / 2 for n in range(N)])
arr3 = np.array([np.min(X[0 : n + 1]) * (n + 2) for n in range(N)])
arr4 = np.array([np.min(X[0 : n + 1]) + np.max(X[0 : n + 1]) for n in range(N)])
arr5 = np.array([np.max(X[0 : n + 1]) * (n + 1) / (n + 2) for n in range(N)])

x_arr = np.arange(1, 10001)
a = np.abs(arr1 - teta)
b = np.abs(arr2 - teta)
c = np.abs(arr3 - teta)
d = np.abs(arr4 - teta)
e = np.abs(arr5 - teta)
```

```python
plt.figure(figsize=(15, 8))
plt.plot(x_arr, a)
plt.plot(x_arr, b)
plt.plot(x_arr, c, 'y')
plt.plot(x_arr, d)
plt.plot(x_arr, e)

plt.ylim([0, 0.5])
plt.xlim([500, 10000])
plt.xticks(np.arange(500, 10001, 500))
plt.title('Task 1')
plt.xlabel('|teta - teta* |')
plt.ylabel('n')
plt.legend(['2<X>', '<X> + X(n)', '(n + 1)*X(1)', 'X(1) + X(n)', '(n + 1)/n * X(n)'], loc = 0, framealpha=0.5)
```
