```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
teta = 1
N = 10000
expon_rv = sts.expon(teta)
X = np.array(expon_rv.rvs(N))
```

```python
#напишем функцию, которая будет генерировать нужную нам выборку, считать оценку и строить график, затем будем просто ее вызывать
def magic(k):
    C = X ** k
    arr2 = (math.factorial(k) / np.array([np.sum(C[0 : n + 1]) / (n + 1) for n in range(N)])) ** (1. / k)
    x_arr = np.arange(1, 10001)
    b = np.abs(arr2 - teta)
    plt.figure(figsize=(15, 4))
    plt.xlim([500, 10000])
    plt.xticks(np.arange(500, 10001, 500))
    plt.plot(x_arr, b)
    plt.title('Task 2')
    plt.xlabel('|teta - teta* |')
    plt.ylabel('n')
    plt.ylim([0, 0.5])
```

```python
magic(5)
```
