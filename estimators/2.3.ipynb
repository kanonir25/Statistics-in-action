```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import scipy.stats as sts
from numpy import var
%matplotlib inline
```

Заметим, что у распределения Парето с параметрами (0, 5) четыре момента будут существовать, а 5-й - нет. Сгенерируем выборку из этого распределения!

```python
N = 10000
pareto_rv = sts.pareto(5)
sample = np.array(pareto_rv.rvs(N))
```

```python
##Построим график ее плотности
x = np.linspace(0,4,100)
pdf = pareto_rv.pdf(x)
plt.figure(figsize=(20, 7))
null = np.zeros(N)
plt.plot(x, pdf)
plt.xlim([0.9, 3])
plt.scatter(sample, null)
plt.ylabel('$pdf(x)$')
plt.xlabel('$x$')
```

```python
##Посчитаем оценку ее дисперсии и построим график
C = sample ** 2 
arr1 = np.array([np.sum(C[0 : n + 1]) / (n + 1) for n in range (N)])
arr2 = np.array([np.sum(sample[0 : n + 1]) / (n + 1) for n in range (N)]) ** 2
stat = arr1 - arr2
varr = [var(sample[0 : n + 1 ]) for n in range(N)]
arr = np.abs(varr - stat)
```

```python
plt.figure(figsize=(25, 8))
plt.plot(range(N), arr)
plt.ylabel('$|S^2 - sigma^2|$')
plt.xlabel('$n$')
```

Проведем аналогичное исследование для выборки из распределения Коши, где построим график оценки дисперсии.

```python
N = 100000
cauchy_rv = sts.cauchy(0, 1)
cauchy = cauchy_rv.rvs(N)
```

```python
C = cauchy ** 2 
arr1 = np.array([np.sum(C[0 : n + 1]) / (n + 1) for n in range (N)])
arr2 = np.array([np.sum(cauchy[0 : n + 1]) / (n + 1) for n in range (N)]) ** 2
stat = arr1 - arr2
```

```python
plt.figure(figsize=(25, 8))
plt.plot(range(N), stat)
plt.title('Dispersion of Cauchy distribution')
plt.ylabel('$sigma^2$')
plt.xlabel('$n$')
```
